```python
# ==============================================================================
# PROJECT: Kaggle Playground Series - S6E2 (Heart Disease)
# MISSION: Final Optimized Submission (Optuna Result: 0.95548)
# ==============================================================================

import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from google.colab import drive

# --- STEP 1: MOUNT & LOAD ---
drive.mount('/content/drive')
TRAIN_PATH = '/content/drive/MyDrive/Nihal Data/kaggle/S6E1 - heart/train.csv'
TEST_PATH = '/content/drive/MyDrive/Nihal Data/kaggle/S6E1 - heart/test.csv'

train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)

# --- STEP 2: DATA PREPARATION ---
target_mapping = {'Absence': 0, 'Presence': 1}
y = train['Heart Disease'].map(target_mapping)
X = train.drop(['id', 'Heart Disease'], axis=1)
X_test = test.drop(['id'], axis=1)

# Categorical Encoding
for col in X.select_dtypes(include=['object']).columns:
    X[col] = X[col].astype('category').cat.codes
    X_test[col] = X_test[col].astype('category').cat.codes

# --- STEP 3: APPLYING YOUR OPTIMIZED PARAMS ---
# These parameters were mathematically derived in Trial 4 to maximize AUC-ROC.
optimized_params = {
    'n_estimators': 3000,
    'learning_rate': 0.0524336233119705,
    'max_depth': 4,
    'subsample': 0.8174286638784904,
    'colsample_bytree': 0.5096944467506977,
    'min_child_weight': 10,
    'tree_method': 'hist',
    'device': 'cuda',
    'eval_metric': 'auc',
    'random_state': 42
}

# --- STEP 4: FINAL TRAINING ---
# Training on 100% of the data ensures the model learns every possible signal.
print("Training the final model with optimized parameters...")
final_model = XGBClassifier(**optimized_params)
final_model.fit(X, y)

# --- STEP 5: GENERATING PREDICTIONS ---
test_probs = final_model.predict_proba(X_test)[:, 1]
submission = pd.DataFrame({'id': test['id'], 'Heart Disease': test_probs})

# Save to the local environment
submission.to_csv('submission_optimized_trial_05.csv', index=False)

print("Process Complete. 'submission_optimized_trial_05.csv' is ready.")
```
